```python
pdData.set_index(pdData.columns[0], inplace=True)
# Вывод названий всех строк (индексов)
pdData.index
```

```python
# Функция для получения уникальных значений в каждом столбце
unique_values_per_column = pdData.apply(pd.Series.unique)
unique_values_per_column
```

```python
from sklearn.preprocessing import LabelEncoder

# Применение Label Encoding к категориальным данным
label_encoders = {}
for column in pdData.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    pdData[column] = le.fit_transform(pdData[column])
    label_encoders[column] = le

pdData.head()
```

```python
for column, le in label_encoders.items():
    print(f"Column: {column}")
    print(dict(zip(le.classes_, le.transform(le.classes_))))
```

```python
def CheckLE(label_encoders, name):
  if name in label_encoders:
    le = label_encoders[name]
    return f"Column: {name} \nMapping: {dict(zip(le.classes_, le.transform(le.classes_)))}"
```

```python
plt.hist(pdData['Age'] / 365)
```

```python
sns.countplot(x=pdData["Sex"]);
print(CheckLE(label_encoders, "Sex"))
```

```python
sns.countplot(x=pdData["Hepatomegaly"]);
print(CheckLE(label_encoders, "Hepatomegaly"))
```

```python
sns.countplot(data=pdData, x="Hepatomegaly", hue='Sex');
print(CheckLE(label_encoders, "Hepatomegaly"))
print(CheckLE(label_encoders, "Sex"))
```

```python
sns.countplot(data=pdData, x="Sex", hue='Status');
print(CheckLE(label_encoders, "Sex"))
print(CheckLE(label_encoders, "Status"))
```

```python
pdData["Age_group"] = pd.cut(x=pdData['Age'] / 365, bins=[0, 30, 40, 50, 60, 70, 100])
plt.xticks(rotation=45)
sns.countplot(data=pdData, hue='Status', x='Age_group');
```
