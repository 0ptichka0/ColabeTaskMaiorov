```python
#Функция суммировния мтриц
def MatrixSum(matrix1: list, matrix2: list):
  if len(matrix1) != len(matrix2) or len(matrix1[0]) != len(matrix2[0]):
    return None

  result = []
  for i in range(len(matrix1)):
    intermediateResult = []
    for j in range(len(matrix1[i])):
      intermediateResult.append(matrix1[i][j] + matrix2[i][j])
    result.append(intermediateResult)
  return result

#Функция разности матриц
def MatrixSub(matrix1: list, matrix2: list):
  if len(matrix1) != len(matrix2) or len(matrix1[0]) != len(matrix2[0]):
    return None

  result = []
  for i in range(len(matrix1)):
    intermediateResult = []
    for j in range(len(matrix1[i])):
      intermediateResult.append(matrix1[i][j] - matrix2[i][j])
    result.append(intermediateResult)
  return result

#Функция перемножения мтриц
def MatrixMult(matrix1: list, matrix2: list):
  if len(matrix1[0]) != len(matrix2):
    return None

  result = []
  for i in range(len(matrix1)):
    intermediateResult = []
    for j in range(len(matrix2[i])):
      elSum = 0
      for l in range(len(matrix2)):
        elSum += matrix1[i][l] * matrix2[l][j]
      intermediateResult.append(elSum)
    result.append(intermediateResult)
  return result

def MatrixStep(matrix: list, step: int):
  result = matrix
  if step == 1:
    return result
  for i in range(step - 1):
    result = MatrixMult(result, matrix)
  return result

def MatrixMultByNumb(matrix: list, numb: float):
  result = []
  for i in range(len(matrix)):
    intermediateResult = []
    for j in range(len(matrix[i])):
      intermediateResult.append(matrix[i][j] * numb)
    result.append(intermediateResult)
  return result
```
